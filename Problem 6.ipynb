```python
import numpy as np
import matplotlib.pyplot as plt
def constant(x,c=1):
    return c*np.ones(len(x))

x_min = -20
x_max = 20
n = 128
x_arr = []
for i in range (n):
    x_arr.append(-20 + (40/n)*i)
x_arr = np.array(x_arr)
plt.plot(x_arr,constant(x_arr),label = "constant function")
plt.legend()

dft_arr = np.fft.fft(constant(x_arr),norm = "ortho")
k_arr =2*np.pi* np.fft.fftfreq(n,(40/n))
ft_arr = np.zeros(len(k_arr))

for i in range(len(k_arr)):
    ft_arr[i]=np.abs((40/n)*np.sqrt(n/(2*np.pi))*np.exp(-(1j)*k_arr[i]*x_min)*(dft_arr[i]))

def pairsort(arr1, arr2):
    pairs = [(arr1[i], arr2[i]) for i in range(len(arr1))]
    sorted_pairs = sorted(pairs, key=lambda x: x[0])
    sorted_arr2 = [pair[1] for pair in sorted_pairs]
    
    return sorted_arr2



plt.plot(np.sort(k_arr), pairsort(k_arr,ft_arr),color = "red",label = "calculated fourier transform")
plt.legend()
plt.show
```
