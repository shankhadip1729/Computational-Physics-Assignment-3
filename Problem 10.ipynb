```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the Gaussian function
def gaussian(x, y):
    return np.exp(- (x**2 + y**2))

# Define the grid
N = 100
x = np.linspace(-5, 5, N)
y = np.linspace(-5, 5, N)
X, Y = np.meshgrid(x, y)

# Compute the Gaussian function
f = gaussian(X, Y)

# Compute the Fourier transform
F = np.fft.fftshift(np.fft.fft2(np.fft.ifftshift(f)))

# Define the frequency grid
kx = np.fft.fftshift(np.fft.fftfreq(N, d=(x[1]-x[0])))
ky = np.fft.fftshift(np.fft.fftfreq(N, d=(y[1]-y[0])))
KX, KY = np.meshgrid(kx, ky)

# Compute the analytical result of the Fourier transform
analytical_F = np.exp(- (KX**2 + KY**2) / 4) / 2

# Create 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the numerical result
ax.plot_surface(KX, KY, np.abs(F), cmap='viridis', edgecolor='none', alpha=0.8, label='Numerical')

# Plot the analytical result
ax.plot_surface(KX, KY, analytical_F, cmap='plasma', edgecolor='none', alpha=0.6, label='Analytical')

ax.set_title('2D Fourier Transform of Gaussian Function')
ax.set_xlabel('kx')
ax.set_ylabel('ky')
ax.set_zlabel('Magnitude')

ax.legend
plt.show()

```
