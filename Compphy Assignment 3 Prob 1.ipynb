```python
import numpy as np
import matplotlib.pyplot as plt
def sync(a):
    if a == 0:
        return(1)
    else:
        return(np.sin(a)/a)
sync = np.vectorize(sync)

x_min = -20
x_max = 20
n = 128
x_arr = []
for i in range (n):
    x_arr.append(-20 + (40/n)*i)
x_arr = np.array(x_arr)
plt.plot(x_arr,sync(x_arr),label = "sync function")
plt.legend()
plt.show()
dft_arr = np.fft.fft(sync(x_arr),norm = "ortho")
k_arr =2*np.pi* np.fft.fftfreq(n,(40/n))
ft_arr = np.zeros(len(k_arr))

for i in range(len(k_arr)):
    ft_arr[i]=np.abs((40/n)*np.sqrt(n/(2*np.pi))*np.exp(-(1j)*k_arr[i]*x_min)*(dft_arr[i]))

def pairsort(arr1, arr2):
    pairs = [(arr1[i], arr2[i]) for i in range(len(arr1))]
    sorted_pairs = sorted(pairs, key=lambda x: x[0])
    sorted_arr2 = [pair[1] for pair in sorted_pairs]
    
    return sorted_arr2



plt.plot(np.sort(k_arr), pairsort(k_arr,ft_arr),label = "calculated fourier transform")
def actual_soln(x):
    if x>-1 and x<1:
        return (np.sqrt(3.14159/2))
    else:
        return 0
actual_soln = np.vectorize(actual_soln)

plt.scatter(k_arr,actual_soln(k_arr),color = "red",label = "analytical fourier transform")
plt.grid
plt.legend()
plt.xlabel("k")
plt.ylabel("fourier transform")
```
